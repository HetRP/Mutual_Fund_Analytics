```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from pathlib import Path

PROCESSED_DIR = Path("../data/processed")
CHART_DIR = Path("../reports/charts")
CHART_DIR.mkdir(parents=True, exist_ok=True)

sns.set_theme(style="whitegrid")
```

```python
fund_master = pd.read_csv(PROCESSED_DIR / "01_fund_master_cleaned.csv")
nav = pd.read_csv(PROCESSED_DIR / "02_nav_history_cleaned.csv")
transactions = pd.read_csv(PROCESSED_DIR / "08_investor_transactions_cleaned.csv")
holdings = pd.read_csv(PROCESSED_DIR / "09_portfolio_holdings_cleaned.csv")
performance = pd.read_csv(PROCESSED_DIR / "07_scheme_performance_cleaned.csv")

nav["date"] = pd.to_datetime(nav["date"])
transactions["transaction_date"] = pd.to_datetime(transactions["transaction_date"])
holdings["portfolio_date"] = pd.to_datetime(holdings["portfolio_date"])

nav = nav.sort_values(["amfi_code", "date"])
```

```python
nav["daily_return"] = nav.groupby("amfi_code")["nav"].pct_change()

returns = nav.dropna(subset=["daily_return"]).copy()

returns.head()
```

```python
var_cvar_results = []

for amfi_code, group in returns.groupby("amfi_code"):
    var_95 = np.percentile(group["daily_return"], 5)

    cvar_95 = group[
        group["daily_return"] <= var_95
    ]["daily_return"].mean()

    var_cvar_results.append({
        "amfi_code": amfi_code,
        "historical_var_95": var_95,
        "historical_cvar_95": cvar_95
    })

var_cvar = pd.DataFrame(var_cvar_results)

var_cvar = var_cvar.merge(
    fund_master[["amfi_code", "scheme_name", "fund_house", "category", "risk_category"]],
    on="amfi_code",
    how="left"
)

var_cvar.to_csv(PROCESSED_DIR / "var_cvar_report.csv", index=False)

var_cvar.sort_values("historical_var_95").head(10)
```

```python
selected_funds = fund_master["amfi_code"].head(5).tolist()

rolling_data = returns[
    returns["amfi_code"].isin(selected_funds)
].copy()

rolling_data["rolling_90d_sharpe"] = rolling_data.groupby("amfi_code")["daily_return"].transform(
    lambda x: (x.rolling(90).mean() / x.rolling(90).std()) * np.sqrt(252)
)

rolling_data = rolling_data.merge(
    fund_master[["amfi_code", "scheme_name"]],
    on="amfi_code",
    how="left"
)

plt.figure(figsize=(14, 7))

sns.lineplot(
    data=rolling_data,
    x="date",
    y="rolling_90d_sharpe",
    hue="scheme_name"
)

plt.title("Rolling 90-Day Sharpe Ratio for 5 Key Funds")
plt.xlabel("Date")
plt.ylabel("Rolling 90-Day Sharpe")
plt.tight_layout()
plt.savefig(CHART_DIR / "rolling_sharpe_chart.png", dpi=300)
plt.show()
```

```python
first_txn = transactions.groupby("investor_id")["transaction_date"].min().reset_index()
first_txn.columns = ["investor_id", "first_transaction_date"]
first_txn["cohort_year"] = first_txn["first_transaction_date"].dt.year

txn_cohort = transactions.merge(first_txn, on="investor_id", how="left")

sip_txn = txn_cohort[txn_cohort["transaction_type"] == "SIP"].copy()

cohort_summary = sip_txn.groupby("cohort_year").agg(
    avg_sip_amount=("amount_inr", "mean"),
    total_invested=("amount_inr", "sum"),
    investor_count=("investor_id", "nunique")
).reset_index()

top_fund = sip_txn.groupby(["cohort_year", "amfi_code"])["amount_inr"].sum().reset_index()
top_fund = top_fund.sort_values(["cohort_year", "amount_inr"], ascending=[True, False])
top_fund = top_fund.groupby("cohort_year").head(1)

top_fund = top_fund.merge(
    fund_master[["amfi_code", "scheme_name"]],
    on="amfi_code",
    how="left"
)

cohort_summary = cohort_summary.merge(
    top_fund[["cohort_year", "scheme_name"]],
    on="cohort_year",
    how="left"
)

cohort_summary.rename(columns={"scheme_name": "top_fund_preference"}, inplace=True)

cohort_summary
```

```python
sip_only = transactions[
    transactions["transaction_type"] == "SIP"
].copy()

sip_counts = sip_only.groupby("investor_id").size().reset_index(name="sip_count")

regular_sip_investors = sip_counts[
    sip_counts["sip_count"] >= 6
]["investor_id"]

sip_regular = sip_only[
    sip_only["investor_id"].isin(regular_sip_investors)
].copy()

sip_regular = sip_regular.sort_values(["investor_id", "transaction_date"])

sip_regular["gap_days"] = sip_regular.groupby("investor_id")["transaction_date"].diff().dt.days

continuity = sip_regular.groupby("investor_id").agg(
    sip_count=("transaction_date", "count"),
    avg_gap_days=("gap_days", "mean"),
    total_sip_amount=("amount_inr", "sum")
).reset_index()

continuity["sip_status"] = np.where(
    continuity["avg_gap_days"] > 35,
    "At-risk",
    "Regular"
)

continuity_rate = continuity["sip_status"].value_counts(normalize=True) * 100

continuity.head()
```

```python
plt.figure(figsize=(7, 5))

sns.countplot(
    data=continuity,
    x="sip_status"
)

plt.title("SIP Continuity Status")
plt.xlabel("SIP Status")
plt.ylabel("Investor Count")
plt.tight_layout()
plt.savefig(CHART_DIR / "sip_continuity_status.png", dpi=300)
plt.show()

continuity_rate
```

```python
def recommend_funds(risk_appetite):
    risk_map = {
        "Low": ["Low", "Moderately Low"],
        "Moderate": ["Moderate"],
        "High": ["High", "Very High"]
    }

    allowed_risks = risk_map.get(risk_appetite)

    if allowed_risks is None:
        print("Please enter risk appetite as Low, Moderate, or High")
        return None

    recommendations = performance[
        performance["risk_grade"].isin(allowed_risks)
    ].copy()

    recommendations = recommendations.sort_values(
        "sharpe_ratio",
        ascending=False
    ).head(3)

    return recommendations[[
        "scheme_name",
        "fund_house",
        "category",
        "risk_grade",
        "return_3yr_pct",
        "sharpe_ratio",
        "expense_ratio_pct",
        "aum_crore"
    ]]

recommend_funds("Moderate")
```

```python
holdings["weight_decimal"] = holdings["weight_pct"] / 100

hhi = holdings.groupby("amfi_code").apply(
    lambda x: np.sum(x["weight_decimal"] ** 2)
).reset_index(name="sector_hhi")

hhi = hhi.merge(
    fund_master[["amfi_code", "scheme_name", "fund_house", "category"]],
    on="amfi_code",
    how="left"
)

hhi = hhi.sort_values("sector_hhi", ascending=False)

hhi.head(10)
```

```python
plt.figure(figsize=(12, 7))

sns.barplot(
    data=hhi.head(15),
    x="sector_hhi",
    y="scheme_name"
)

plt.title("Top 15 Funds by Sector Concentration HHI")
plt.xlabel("Sector HHI")
plt.ylabel("Scheme Name")
plt.tight_layout()
plt.savefig(CHART_DIR / "sector_hhi_concentration.png", dpi=300)
plt.show()
```

```python
## Advanced Insight 4
# SIP investors with average payment gaps above 35 days are marked as at-risk because they may discontinue regular investing.
```

```python
## Advanced Insight 5
# Funds with high sector HHI are more concentrated and may carry higher sector-specific risk.
```
